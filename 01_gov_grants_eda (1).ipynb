## **3.2 EDA: What is the number of grants over time?**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert posted date to posted year
grants['posted_year'] = pd.to_datetime(grants['posted_date']).dt.year

# Count and sort grants by posted year
grants_by_year = grants['posted_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
grants_by_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Grants')
plt.title('Number of Grant Opportunities Posted Over Time')
plt.grid(True, alpha=0.3)
plt.xticks(grants_by_year.index)
plt.show()
```

**Analysis:** As seen in the line chart above, the number of grants have drastically picked up over time-- specifically from 2007 to 2010. The peak of number of grants posted was in 2016 (over 6,000 grants). The number has, overall, lowered since. It is important to note that the grants could have risen because of an increase in tracking them into this dataset.

## **3.3 EDA: What is the value of grants over time?**

```python
funding_by_year = grants.groupby('posted_year')['estimated_funding'].sum()

plt.figure(figsize=(12,6))
funding_by_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Total Estimated Funding ($100 Billions)')
plt.title('Value of Grant Opportunities Posted Over Time')
plt.grid(True, alpha=0.3)
plt.xticks(funding_by_year.index)
plt.show()
```

```python
# Verify total funding in 2022
funding_2022 = grants[grants['posted_year'] == 2022]['estimated_funding'].sum()
print(f"Total funding in 2022: ${funding_2022:,.0f}")
```

**Follow-up question- what were the top 5 agencies that this high funding went to in 2022?**

```python
grants_2022 = grants[grants['posted_year'] == 2022]
top_5_agencies_2022 = grants_2022.groupby('agency_name')['estimated_funding'].sum().sort_values(ascending=False).head(5)

print("Top 5 agencies by funding in 2022:", top_5_agencies_2022)
```

**Analysis:** As we can see above, the y values are being taken to the 10^11 making the values of the y axis be in 100 billion dollars.

We see a large spike in 2022 of around 300 billion dollars of funding. After many slow steady years besides a small 2009 increase, 2019-2022 see a massive rise. 2023 then falls down to around 150 billion dollars. We wonder if this could do with post-crisis moments- like the 2008 financial crisis and COVID. 2022's jump could also be connected with the student loan forgiveness program.

It is again important to note that the grants could have risen because of an increase in tracking them into this dataset.

## **3.6 EDA: What eligibility types are most common, and how do they co-occur?**

Eligibility flags define who can apply for each grant  nonprofits, universities, tribal governments, individuals, etc. Understanding which applicant types dominate and which combinations frequently co-occur helps uncover natural groupings of grants, targeted funding patterns, and structural biases in federal funding. This is ideal for unsupervised learning tasks like segmentation, association analysis, and clustering.


```python
# Identify eligibility columns
elig_cols = [col for col in details.columns if col.startswith('eligibility_')]

# Total count per eligibility type
eligibility_counts = details[elig_cols].sum().sort_values(ascending=False)

# Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
eligibility_counts.plot(kind='bar')
plt.title("Frequency of Eligibility Types Across Grant Opportunities")
plt.ylabel("Number of Grants")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Correlation matrix to explore co-occurrence
eligibility_corr = details[elig_cols].corr()
```

**Analysis:** Most grants are open to institutions like universities and nonprofits, while individuals and for-profit entities are rarely eligible. Eligibility_others dominates, suggesting broad or undefined criteria. These patterns reveal structural biases and support clustering grants by applicant type for segmentation or recommendation tasks.

## **3.7 EDA: Which grant categories are most common, and how do they overlap?**

Each grant is tagged with one or more thematic categories (e.g., health, science, education, environment). Understanding which categories dominate and which combinations frequently co-occur reveals funding priorities, natural clusters, and multi-domain intersections. This is ideal for unsupervised learning tasks like topic segmentation, association analysis, and multilabel clustering.


```python
# Identify category columns but EXCLUDE the explanation column
cat_cols = [col for col in details.columns
            if col.startswith('category_') and col != 'category_explanation']

# Total count per category
category_counts = details[cat_cols].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12,6))
category_counts.plot(kind='bar')
plt.title("Frequency of Grant Categories")
plt.ylabel("Number of Grants")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

category_corr = details[cat_cols].corr()
```

**Analysis:** Health, science, education, and environment are the most frequent grant categories, indicating strong federal focus in these domains. Categories like arts, humanities, and information services are underrepresented, suggesting niche or underserved areas. These patterns support clustering grants by thematic focus and exploring multi-category overlaps for segmentation.

## **3.8 EDA: How much do grant descriptions vary in length?**


Grant descriptions contain rich, unstructured text that communicates program goals, eligibility, and scope. Analyzing their length reveals narrative complexity, variation in detail, and potential for clustering grants by verbosity or structure. This is ideal for unsupervised tasks like representation learning, topic modeling, and embedding-based similarity search.


```python
# Calculate description length
details['description_length'] = details['description'].str.len()

# Box plot to show distribution
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x=details['description_length'])
plt.title("Distribution of Grant Description Lengths")
plt.xlabel("Description Length (characters)")
plt.tight_layout()
plt.show()

# Summary stats
details['description_length'].describe()
```

**Analysis:** Grant descriptions vary widely in length, with a median around 985 characters and some exceeding 16,000. This high variance suggests differences in narrative complexity and program scope. Such variation supports clustering grants by verbosity and applying NLP techniques for representation learning or topic modeling.

## **3.9 EDA: Are there seasonal patterns in grant postings?**


Understanding when grants are posted helps identify cyclical funding behavior, fiscal year effects, and agency timing strategies. This insight supports anomaly detection, time-based segmentation, and forecasting.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert posted_date to datetime and extract month
details['posted_date'] = pd.to_datetime(details['posted_date'], errors='coerce')
details['posted_month'] = details['posted_date'].dt.month

# Count grants by month
monthly_counts = details['posted_month'].value_counts().sort_index()

# Prepare DataFrame for Seaborn compatibility
monthly_df = pd.DataFrame({
    'Month': monthly_counts.index,
    'Count': monthly_counts.values
})

# Plot with hue explicitly set to 'Month'
plt.figure(figsize=(10, 6))
sns.barplot(data=monthly_df, x='Month', y='Count', hue='Month', palette="coolwarm", legend=False)
plt.title("Monthly Distribution of Grant Postings")
plt.xlabel("Month")
plt.ylabel("Number of Grants Posted")
plt.xticks(ticks=range(0,12), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                                      "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.tight_layout()
plt.show()
```

**Analysis:** Grant postings peak in August and September, aligning with the end of the federal fiscal year. Postings drop sharply in October through December. This seasonal pattern suggests cyclical funding behavior and supports time-based segmentation or forecasting models.

## **3.11 EDA: Crisis-Driven Grant Analysis: COVID-19, Ukraine, and School Funding**

Our dataset likely contains grants related to COVID-19, Ukraine war support, and school funding, especially from 20202023. These themes were major drivers of federal spending, with billions allocated through emergency acts and education recovery programs. You can surface them using keyword filters in the  or  fields.


```python
school_related = details[details['description'].str.contains("school|education|students|K-12", case=False, na=False)]
print(f"School-related grants: {len(school_related)}")
```

```python
covid_related = details[details['description'].str.contains("COVID|pandemic|coronavirus|CARES|ARP", case=False, na=False)]
print(f"COVID-related grants: {len(covid_related)}")
```

```python
ukraine_related = details[details['description'].str.contains("Ukraine|Russia|conflict|refugee|security", case=False, na=False)]
print(f"Ukraine-related grants: {len(ukraine_related)}")
```

**1. Compare funding totals by theme**


```python
print("School funding total:", school_related['estimated_total_program_funding'].sum())
print("COVID funding total:", covid_related['estimated_total_program_funding'].sum())
print("Ukraine-related funding total:", ukraine_related['estimated_total_program_funding'].sum())
```

**2. Plot grant counts over time by theme**

```python
# Safely create new columns by copying the slices first
school_related = school_related.copy()
covid_related = covid_related.copy()
ukraine_related = ukraine_related.copy()

# Extract year from posted_date
school_related['year'] = pd.to_datetime(school_related['posted_date'], errors='coerce').dt.year
covid_related['year'] = pd.to_datetime(covid_related['posted_date'], errors='coerce').dt.year
ukraine_related['year'] = pd.to_datetime(ukraine_related['posted_date'], errors='coerce').dt.year

# Count per year
school_counts = school_related['year'].value_counts().sort_index()
covid_counts = covid_related['year'].value_counts().sort_index()
ukraine_counts = ukraine_related['year'].value_counts().sort_index()

# Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(school_counts.index, school_counts.values, label='School Grants')
plt.plot(covid_counts.index, covid_counts.values, label='COVID Grants')
plt.plot(ukraine_counts.index, ukraine_counts.values, label='Ukraine Grants')
plt.title("Grant Counts Over Time by Theme")
plt.xlabel("Year")
plt.ylabel("Number of Grants")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Analysis:**

To explore how global events shaped federal grant activity, we filtered the dataset using keyword searches in the `description` field. This surfaced grants related to:

- **School/Education**: 372 grants  
- **COVID-19**: 46 grants  
- **Ukraine/Russia Conflict**: 125 grants  

These counts reflect major policy responses, especially in education recovery and international aid.
